# Define the states and their probability: the transition matrix

```python
# The statespace
states = ["Hamburger","pizza","hotdog"]

# Possible sequences of events
transitionName = [["HH","Hp","Hh"],["pH","pp","ph"],["hH","hp","hh"]]

# Probabilities matrix (transition matrix)
transitionMatrix = [[0.2,0.6,0.2],[0.3,0,0.7],[0.5,0,0.5]]
```

# Make sure the probabilities sum up to 3

```python
if sum(transitionMatrix[0])+sum(transitionMatrix[1])+sum(transitionMatrix[1]) != 3:
    print("Somewhere, something went wrong. Transition matrix, perhaps?")
else: print("All is gonna be okay, you should move on!! ;)")
```

```python
# A function that implements the Markov model to forecast the state/mood.
def activity_forecast(days):
    # Choose the starting state
    activityToday = "Hamburger"
    print("Start state: " + activityToday)
    # Shall store the sequence of states taken. So, this only has the starting state for now.
    activityList = [activityToday]
    i = 0
    # To calculate the probability of the activityList
    prob = 1
    countH=0
    counth=0
    countp=0
    while i != days:
        if activityToday == "Hamburger":
            countH+=1
            change = np.random.choice(transitionName[0],replace=True,p=transitionMatrix[0])
            if change == "HH":
                prob = prob * 0.2
                activityList.append("Hamburger")
                pass
            elif change == "Hp":
                prob = prob * 0.6
                activityToday = "pizza"
                activityList.append("pizza")
            else:
                prob = prob * 0.2
                activityToday = "hotdog"
                activityList.append("hotdog")
        elif activityToday == "pizza":
            countp+=1
            change = np.random.choice(transitionName[1],replace=True,p=transitionMatrix[1])
            if change == "pH":
                prob = prob * 0.3
                activityToday = "Hamburger"
                activityList.append("Hamburger")
            elif change == "pp":
                prob = prob * 0
                activityList.append("pizza")
                pass
            elif change=='ph':
                prob = prob * 0.7
                activityToday = "hotdog"
                activityList.append("hotdog")
        elif activityToday == "hotdog":
            counth+=1
            change = np.random.choice(transitionName[2],replace=True,p=transitionMatrix[2])
            if change == "hH":
                prob = prob * 0.5
                activityToday = "Hamburger"
                activityList.append("Hamburger")
            elif change == "hp":
                prob = prob * 0
                activityToday = "pizza"
                activityList.append("pizza")
            elif change=='hh':
                prob = prob * 0.5
                activityList.append("hotdog")
                pass
        i += 1  
    print("Possible states: " + str(activityList))
    print("End state after "+ str(days) + " days: " + activityToday)
    print("Probability of Hamburger for the next 10000 days:",(countH/10000)*100,"%")
    print("Probability of pizza for the next 10000 days:",(countp/10000)*100,"%")
    print("Probability of hotdog for the next 10000 days:",(counth/10000)*100,"%")

# Function that forecasts the possible state for the next 10000 days
activity_forecast(10000)
```

# Split the text file into single words

```python
words = data.split()
words
```

# Generating pairs. To do this, we use a special generator command: yield to give us all the pairs of words

```python
def make_pairs(words):
    for i in range(len(words)-1):
        yield (words[i], words[i+1])
pairs = make_pairs(words)
```

# Using these pairs, we compose a dictionary of strings, where the first word is indicated and everything that can come after it.

```python
word_dict = {}
for word_1, word_2 in pairs:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
```

# Pick some random word to kick off the chain

```python
first_word = np.random.choice(words)
while first_word.islower():#require the first word to be capitalized,
    first_word = np.random.choice(words)
chain = [first_word]
n_words = 100#number of max words in text
```

# Every word in the chain is sampled randomly from the list of words

```python
for i in range(n_words):
    chain.append(np.random.choice(word_dict[chain[-1]]))
' '.join(chain)# returns the chain as a string:
```

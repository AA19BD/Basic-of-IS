```python
import numpy as np
from sklearn import datasets
import math
```

```python
iris = datasets.load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
minn = iris.data.min(axis=0)
print(minn)
```

```python
mixx = iris.data.max(axis=0)
print(mixx)
```

```python
dataset = np.floor((iris.data-minn)*10/(mixx-minn)).astype(int)
```

```python
print(dataset)
```

```python
res = np.zeros((11,11,11,11,3))
```

```python
res
```

```python
for i in range(dataset.shape[0]):
    res[dataset[i][0],dataset[i][1],dataset[i][2],dataset[i][3],iris.target[i]] +=1
    
```

```python
res = res/res.sum()
```

```python
mi = 0
for x in range(0,11):
    for y in range(0,3):
        pxy = res[:,x,:,:,y].sum()
        px=res[:,x,:,:,:].sum()
        py=res[:,:,:,:,y].sum()
        if pxy>0 and px>0 and py>0:
            mi += pxy*math.log2(pxy/(px*py))
print(mi)
```

```python
import re 
import math
import nltk
# nltk.download('punkt')
# from nltk.tokenize import *
with open('jest_vek.txt','r',encoding="utf-8") as fh:
    data = fh.read()
    data = data.lower()
    lines = data.split("\n")
newLines = []
for line in lines:
    my_sym=re.findall('[А-Яа-я]',line)
    for i in my_sym:
        newLines.append(i)
# words=nltk.word_tokenize(data)
```

```python
def ConditionalEntropy(words):
    word=Words(words)
    pair=Pairs(words)
    counts3={}
    sum=0
    for i in pair:
        p_x=word[i[0]]
        num=(pair[i]*math.log2(p_x))-(pair[i]*math.log2(pair[i]))
        counts3[i]=num
    for i in counts3:
        sum+=counts3[i]
#         print(i,"",counts3[i])
    return sum

def Words(words):
    counts={}
    for i in words:
        counts[i]=counts.get(i,0)+1
    for i in counts:
        counts[i]/=len(words)
    return counts
def Pairs(words):
    counts1={}
    for i in range(len(words)-1):
        a=(words[i],words[i+1])
        counts1[a]=counts1.get(a,0)+1
    counts2={}
    for i in counts1:
        p_xy=counts1[i]/len(words)
        counts2[i]=p_xy
    return counts2

```

```python
def count_word_frequency(words):
    freg1=dict()
    freg2=dict()
    counts1 =dict()
    bgs1 = [(words[i]) for i in range(0,len(words)-1)]
    bgs2 = [(words[i+1]) for i in range(0,len(words)-1)]
    bgs = [(words[i],words[i+1]) for i in range(0,len(words)-1)]
    sum=0
    for i in bgs1:
        freg1[i]=freg1.get(i,0)+1
        
    for i in bgs2:
        freg2[i]=freg2.get(i,0)+1
        
    for word in bgs:#count_bigram_frequency(P_xy)
        counts1[word]=counts1.get(word,0)+1
     
    PX=[]
    PY=[]
    P_XY=[]
    for v in (freg1.values()):#probability of px
        px=v/len(words)
        PX.append(px)
#         print((px))


    for v in (freg2.values()):#probability of py
        py=v/len(words)  
        PY.append(py)
#       print(k,py)

    for v in (counts1.values()):#probability of p(x,y)
        p_xy=v/len(bgs)
        P_XY.append(p_xy)
#         print(p_xy)

    res=[]
    for num1, num2 in zip(PX, PY):
        res.append(num1 * num2)
#     print(res)
    res1=[]
    for num2, num3 in zip(res, P_XY):#pmi
        h=(math.log2((num3/num2)))
        res1.append(h)
#     print(res1)

   
    

        
 
   

    





    




            
        
        
        

    
        
        
count_word_frequency(words)
```
